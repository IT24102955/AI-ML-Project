```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data_df=pd.read_csv('https://raw.githubusercontent.com/IT24102955/AI-ML-Project/refs/heads/main/data/raw/IT24104218.csv')
```

##Feature Engineering

```python
ax = data_df.hist(
    bins=40,
    figsize=(20, 12),
    color='#4C72B0',
    edgecolor='white',
    grid=False
)

for a in ax.ravel():

    a.set_facecolor("#fdfdfd")


    a.grid(True, linestyle=':', alpha=0.6)


    a.tick_params(axis='x', labelsize=11, colors="#333333")
    a.tick_params(axis='y', labelsize=11, colors="#333333")


    a.set_title(a.get_title(), fontsize=13, fontweight='semibold', color="#444444")


plt.suptitle("Feature Distributions", fontsize=22, fontweight='bold', color="#222222", y=1.05)


plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

As we mentioned above for our prediction model features like Mobile_phone,work_phone,phone,EMAIL_ID so we are going to drop those features

```python
# dropping the above mentioned features
data_df.drop(labels=['Mobile_phone','Work_Phone','Phone','EMAIL_ID'],axis=1,inplace=True)
```

```python
# Feature Distribution after dropping the mentioned features
ax = data_df.hist(
    bins=40,
    figsize=(20, 12),
    color='#4C72B0',
    edgecolor='white',
    grid=False
)

for a in ax.ravel():

    a.set_facecolor("#fdfdfd")


    a.grid(True, linestyle=':', alpha=0.6)


    a.tick_params(axis='x', labelsize=11, colors="#333333")
    a.tick_params(axis='y', labelsize=11, colors="#333333")


    a.set_title(a.get_title(), fontsize=13, fontweight='semibold', color="#444444")


plt.suptitle("Feature Distributions", fontsize=22, fontweight='bold', color="#222222", y=1.05)


plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

```python
# feature correlation matrix for numerical data
numerical_features=data_df.select_dtypes(include=[np.number]).columns.tolist()
correlation_matrix = data_df[numerical_features].corr()
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Feature Correlation Matrix')
plt.show()
```

here features Family_Members and CHILDERN have correlation score 0.89 that means they contain same information most of the time so we can choose one of these 2 features to keep and drop the other one here we decided to drop the CHILDREN column

```python
# dropping the children column
data_df.drop('CHILDREN',axis=1,inplace=True)
```

```python
# correlation matrix after dropping the CHILDREN feaature
numerical_features=data_df.select_dtypes(include=[np.number]).columns.tolist()
correlation_matrix = data_df[numerical_features].corr()
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Feature Correlation Matrix')
plt.show()
```
